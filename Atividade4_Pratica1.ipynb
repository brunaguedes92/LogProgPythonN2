2. Crie um gráfico de série mostrando os casos acumulados no Brasil. Aproveite para
inserir informações descritivas sobre o gráfico, determinando o título, os eixos e o
tamanho.


```python
dados_covid_brazil = dados_covid.query('Country == "Brazil"')
```

```python
plt.figure(figsize=(8,3))
plt.plot(
    dados_covid_brazil['Date_reported'], dados_covid_brazil['Cumulative_cases'], label = 'Brasil'
)
plt.title('Casos Acumulados de Covid-19 no Brasil')
plt.xlabel('Data')
plt.ylabel('Casos Acumulados')
plt.ticklabel_format(style='plain', axis='y')
plt.legend()
```

3. Faça um gráfico de visualização acerca do número total de casos no Brasil e na Índia.


```python
dados_covid_india = dados_covid.query('Country == "India"')
```

```python
plt.figure(figsize=(8,3))
plt.plot(
    dados_covid_brazil['Date_reported'], dados_covid_brazil['Cumulative_cases'], label = 'Brasil'
)
plt.plot(
    dados_covid_india['Date_reported'], dados_covid_india['Cumulative_cases'], label = 'India'
)
plt.title('Casos Acumulados de Covid-19 no Brasil')
plt.xlabel('Data')
plt.ylabel('Casos Acumulados')
plt.ticklabel_format(style='plain', axis='y')
plt.legend()
```

5. Usando a Biblioteca plotly.express, visualize os dados de COVID-19 de todos os
países, contidos no arquivo estudado, que considera o ano de 2020.

```python
import plotly.express as px
```

```python
from plotly.graph_objs import XAxis
fig = px.line(dados_covid, x='Date_reported', y='Cumulative_cases', title='Dados de COVID-19 no mundo', color = 'Country')
fig.update_layout(
    font_family = 'Arial', font_size = 14, font_color = 'grey', title_font_color='black', title_font_size = 40, xaxis_title = 'Data', yaxis_title = 'Número de casos acumulados')
fig.show()
```
